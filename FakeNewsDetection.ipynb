# Data Preprocessing

```python
news_dataset = pd.read_csv('/content/train.csv')
```

Counting the number of missing values in dataset

```python
news_dataset = news_dataset.fillna('')
```

```python
news_dataset['content'] = news_dataset['author']+' '+news_dataset['title']
```

```python
print(news_dataset['content'])
```

Separating the data and the content. label is removed and column is removed

```python
X = news_dataset.drop(columns='label', axis=1)
Y = news_dataset['label']
```

```python
print(X)
print(Y)
```

Stemming- the process of reducing a word to its root word removing all the prefix and suffixes only keeping the root word

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
news_dataset['content'] = news_dataset['content'].apply(stemming)
```

```python
print(news_dataset['content'])
```

```python
X = news_dataset['content'].values
Y = news_dataset['label'].values
```

```python
print(X)
print(Y)
```

```python
Y.shape
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(X)

X = vectorizer.transform(X)
```

```python
print(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score of the training data : ', testing_data_accuracy)
```
